### Naming and using variable:
Guidelines:
* * *
-- Variable name must start with a letter or underscore

--It can contain letters, digits, and underscores only

--Variable name must not start with a number

--Keywords of the programming language cannot be used

--Use meaningful and descriptive names

--Follow naming conventions (camelCase / snake_case)

--Declare variables before use

--Avoid confusing names like l, O, x

`Upper()`

### Definition and Usage

upper() is a string method in Python that converts all lowercase letters in a string to uppercase.
It does not change numbers, spaces, or special characters.

* * *

```python
"PYTHON".upper()
```

* * *

`title()`

### Definition and Usage
title() is a string method in Python that converts a string into title case, meaning:

-->The first letter of each word becomes uppercase

-->The remaining letters become lowercase

```python
"python programming".title()
```
