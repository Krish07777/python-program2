## 1) <span style="color: rgb(123, 23, 162);">Case Conversion Methods</span>

* * *

`lower()`

### Definition and Usage

Return a copy of the string with all the cased characters converted to lowercase. For example:

* * *

```python
"python".lower()
```

`Upper()`

### Definition and Usage

upper() is a string method in Python that converts all lowercase letters in a string to uppercase.
It does not change numbers, spaces, or special characters.

* * *

```python
"PYTHON".upper()
```

* * *

`title()`

### Definition and Usage
title() is a string method in Python that converts a string into title case, meaning:

-->The first letter of each word becomes uppercase

-->The remaining letters become lowercase


* * *

```python
"python programming".title()
```

* * *

`capitalize()`

### Definition and Usage

The `capitalize()` Return a copy of the string with its first character capitalized and the rest lowercased.

 *The first character is now put into titlecase rather than uppercase. This means that characters like digraphs will only have their first letter capitalized, instead of the full character.*

* * *

```python
"python".capitalize()
```

***
`swapcase()`

### Definition and Usage

swapcase() is a Python string method that converts uppercase letters to lowercase and lowercase letters to uppercase. Other characters (digits, symbols, spaces) remain unchanged.

Usage:
It is used when you want to invert the letter case of a string.
* * *

```python
"PyThOn".swapcase()
```

`casefold()` (stronger lower)

### Definition and Usage

Definition:
casefold() is a Python string method that returns a case-folded copy of the string. It is used for caseless string matching, where all case differences are removed.

Usage:
It converts a string to a form suitable for case-insensitive comparisons.
casefold() is more aggressive than lower() and handles special characters correctly.

Why casefold() is stronger than lower():

lower() only converts characters to lowercase

casefold() removes all case distinctions, including special language rules

```python
"GERMAN ß".casefold() 
```

### Definition and Usage

Definition:
Removing spaces (cleanup) means eliminating unnecessary spaces from a string to make the text clean and properly formatted.

Usage:
It is used to clean user input, format text, and prepare data for comparison or storage by removing extra spaces at the beginning, end, or inside a string.

```python
"  hello  ".strip()   # 'hello'
```

`lstrip(chars=None)`

### Definition and Usage

Definition:
Removing spaces (cleanup) means eliminating unnecessary spaces from a string to make the text clean and properly formatted.

Usage:
It is used to clean user input, format text, and prepare data for comparison or storage by removing extra spaces at the beginning, end, or inside a string.

* * *

```python
"hello   ".rstrip()   # 'hello'
```

* * *

`rstrip(chars=None)`

### Definition and Usage

Return a copy of the string with trailing characters removed. The *chars* argument is a string specifying the set of characters to be removed. If omitted or `None`, the *chars* argument defaults to removing whitespace. The *chars* argument is not a suffix; rather, all combinations of its values are stripped:

* * *

```python
"hello   ".rstrip()   # 'hello'
```
