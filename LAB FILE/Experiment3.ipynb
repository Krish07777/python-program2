1. Check whether a given number is divisible by 3 and 5 both. 

```python
n = int(input("Enter number: "))

if n % 3 == 0 and n % 5 == 0:
    print("Divisible by 3 and 5")
else:
    print("Not divisible by both")
```

2. Check whether a given number is multiple of five or not. 

```python
n = int(input("Enter number: "))

if n % 5 == 0:
    print("Multiple of 5")
else:
    print("Not multiple of 5")
```

3. Find the greatest among two numbers. If numbers are equal than print “numbers are equal”.

```python
a = int(input("Enter first: "))
b = int(input("Enter second: "))

if a > b:
    print("Greatest:", a)
elif b > a:
    print("Greatest:", b)
else:
    print("Numbers are equal")
```

5. Check whether the quadratic equation has real roots or imaginary roots. Display the roots.  

```python
import math

a = float(input("a: "))
b = float(input("b: "))
c = float(input("c: "))

d = b*b - 4*a*c

if d > 0:
    r1 = (-b + math.sqrt(d)) / (2*a)
    r2 = (-b - math.sqrt(d)) / (2*a)
    print("Real roots:", r1, r2)

elif d == 0:
    r = -b / (2*a)
    print("Equal roots:", r)

else:
    print("Imaginary roots")
```

6. Find whether a given year is a leap year or not.  

```python
y = int(input("Enter year: "))

if (y % 400 == 0) or (y % 4 == 0 and y % 100 != 0):
    print("Leap Year")
else:
    print("Not Leap Year")
```
