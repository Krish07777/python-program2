# DOCSTRING EXAMPLE

```python
def add_numbers(a, b):
    """
    This function takes two numbers as input
    and returns their sum.
    """
    return a + b

print(add_numbers(5, 3))
```

```python
def multiply(x, y):
    """    Multiply two numbers.

    Parameters:
    x (int or float): First number
    y (int or float): Second number

    Returns:
    int or float: Product of x and y
    """

    return x * y
print(multiply(5,6))
```

# RECUSIVE FUNCTION EXAMPLE

```python
def factorial(n):
    # Base Case
    if n == 0 or n == 1:
        return 1
    # Recursive Case
    else:
        return n * factorial(n - 1)

print(factorial(5))
```

```python
def print_numbers(n):
    if n == 0:
        return
    print_numbers(n - 1)
    print(n)

print_numbers(10)
```

```python
l1 = ["Stu1", "Stu2", "Stu3", "Stu4", "Stu5"]
l2 = ["Stu6", "Stu7"]
pairs = set()
for i in l1:
    for j in l2:
        pairs.add((i, j))

print("All Possible Pairs:")
for pair in pairs:
    print(pair)
print("Total number of pairs:", len(pairs))

```

```python
square = lambda x: x * x

print(square(5))
```

```python
def dot_product(vec1, vec2):
    # Ensure both vectors have the same length
    if len(vec1) != len(vec2):
        raise ValueError("Vectors must be of the same length")
    
    result = 0
    for i in range(len(vec1)):
        result += vec1[i] * vec2[i]
    return result

# Example usage
v1 = [1, 2, 3]
v2 = [4, 5, 6]
print("Dot Product:", dot_product(v1, v2))
# Output: Dot Product: 32
```
