```python
class Car:
    """
    A class to represent a Car and estimate its mileage.
    """

    def __init__(self, brand: str, model: str, engine_l: float, weight_kg: float, tyre_in: float):
        # Attributes
        self.brand = brand
        self.model = model
        self.engine_l = engine_l
        self.weight_kg = weight_kg
        self.tyre_in = tyre_in


    def dummy_mileage_kmpl(self) -> float:
        """
        Dummy (toy) mileage estimator in km/L.
        """
        mileage = 28 - 4 * self.engine_l - 0.006 * (self.weight_kg - 1000) - 0.7 * (self.tyre_in - 15)
        return max(5.0, mileage)  

    
    def city_highway_kmpl(self) -> tuple:
        """
        Returns (base, city, highway) mileage in km/L.
        """
        base = self.dummy_mileage_kmpl()
        city = 0.85 * base
        highway = 1.10 * base
        return base, city, highway

    def display_info(self):
        base, city, highway = self.city_highway_kmpl()
        print(f"\nCar: {self.brand} {self.model}")
        print(f"Engine: {self.engine_l}L")
        print(f"Weight: {self.weight_kg} kg")
        print(f"Tyre Size: {self.tyre_in} inches")
        print(f"Base Mileage   : {base:.2f} km/L")
        print(f"City Mileage   : {city:.2f} km/L")
        print(f"Highway Mileage: {highway:.2f} km/L")



if __name__ == "__main__":
    car1 = Car("Toyota", "Glanz", 1.5, 120, 16)
    car1.display_info()

```
